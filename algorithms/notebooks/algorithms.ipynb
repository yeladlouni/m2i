L'algorithmique permet de mettre en place des algorithmes pour résoudre un problème spécifique et d'analyser le coût de chaque algorithme afin de choisir le plus performant.

Lors de l'écriture des algorithmes, on fait abstraction en général du langage de programmation utilisé.

L'agorithmique se base sur les structures de données qui sont les éléments utilisés afin de stocker les données et sont généralement définies en utilisant les classes.

Parmi les structures de données "data structures" les plus utilisées, on trouve les tableaux "arrays", les listes y compris les listes chaînées "linked lists", les dictionnaires ou tableaux associatifs, les queues "Queues" FIFO et les piles "stacks" LIFO.

FIFO: first in first out
LIFO: last in first out

# Structures de données

**Queue**

Une queue est une structure de données utilisée pour stocker et traiter les éléments dans une approche FIFO "first in first out", le premier élément est le premier traité.

**Pile**

Une pile "stack" permet de stocker les éléments et les traiter selon une approche LIFO, le dernier élément est le premier traité.

Le package **collections** de python contient plusieurs structures de données y compris **deque** qui permet de mettre en place les structures de données queue et pile.

```
from collections import deque
```

```
q = deque()
```

```
q.append('item1')
q.append('item2')
q.append('item3')
q.append('item4')
```

```
print(q)
```

Pour afficher les éléments selon l'approche FIFO "queue":

```
for item in q:
  print(item)
```

Pour afficher les éléments selon l'approche LIFO "stack":

```
while q:
  print(q.pop())
```

**Bubble sort**

C'est un algorithme qui permet de trier un ensemble de données avec une complexité de $\mathcal{O} = n²$

```
def bubble_sort(numbers):
  n = len(numbers)
  for i in range(n - 1):
    for j in range(0, n-i-1):
      if numbers[j] > numbers[j+1]:
        numbers[j], numbers[j+1] = numbers[j+1], numbers[j]
```

**Fonction récursive**

Une fonction récursive est une fonction qui fait appel à elle même.

Exemple: Pour calculer $n! = n * (n-1)....1$

On peut calculer le n! en utilisant une fonction itérative:

```
def factorial(n):
  value = 1
  for i in range(1, n+1):
    value = value * i

  return value
```

```
factorial(3)
```

Et si on utilisait une fonction récursive:

```
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)
```

```
factorial(3)
```

**Merge sort**

Merge sort est un algorithme de tri qui utilise une fonction récursive afin de trier une collection de valeurs, sa complexité algorithmique est de $\mathcal{O} = nlog(n)$

```
def merge(a, b):
  index_a = 0
  index_b = 0
  
  c = []

  while index_a < len(a) and index_b < len(b):
    if a[index_a] <= b[index_b]:
      c.append(a[index_a])
      index_a += 1
    else:
      c.append(b[index_b])
      index_b += 1
  c.extend(a[index_a:])
  c.extend(b[index_b:])

  return c
```

```
def merge_sort(numbers):
  if len(numbers) <= 1:
    return numbers[:len(numbers)]

  m = len(numbers) // 2  # prendre le floor: valeur entière inférieure

  list1 = numbers[:m]
  list2 = numbers[m+1:]
  
  mylist = merge(merge_sort(list1), merge_sort(list2))
  
  return mylist
```

```
sorted_numbers = merge_sort(numbers)
```
